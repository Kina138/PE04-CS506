## Part 0: Import libraries and load dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('PE04_winereview.csv', encoding='latin')
```

## Part 1: Bar chart of 'points' using value_counts()

```python
# Count the frequency of each score in the "points" column
point_counts = df['points'].value_counts().sort_index()

# Plot the value counts as a bar chart
point_counts.plot(kind='bar', figsize=(10, 5))

# Set chart title and axis labels
plt.title("Wine Review Score")
plt.xlabel("Points")
plt.ylabel("Frequency")

# Add horizontal grid lines for readability
plt.grid(axis='y')

# Ensure layout fits without clipping
plt.tight_layout()

# Display the plot
plt.show()
```

## Part 2: Histogram of 'points' with bin sizes 10 and 3

```python
# Drop missing values in "points" column for plotting
points_clean = df['points'].dropna()

# Histogram with bin size = 10
plt.figure(figsize=(10, 5))
plt.hist(points_clean, bins=range(0, 101, 10), edgecolor='black')

# Set title and axis labels
plt.title("Wine Score Histogram (Bin Size 10)")
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.xlim(75, 105) 
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Histogram with bin size = 3
plt.figure(figsize=(10, 5))
plt.hist(points_clean, bins=range(0, 101, 3), edgecolor='black')

# Set title and axis labels
plt.title("Wine Score Histogram (Bin Size 3)")
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.xlim(75, 105) 
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Part 3: Histogram of 'price' with bin size 5

```python
# Convert price column to numeric, invalid values will be set to NaN
price_clean = pd.to_numeric(df['price'], errors='coerce').dropna()

# Create histogram for price with bin size 5
plt.figure(figsize=(10, 5))
plt.hist(price_clean, bins=range(0, int(price_clean.max()) + 5, 5), edgecolor='black')

# Set title and axis labels
plt.title("Wine Price Histogram (Bin Size 5)")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.xlim(0, 550) 
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
